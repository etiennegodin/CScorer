```python
def get_coordinates(polygon:dict) -> str:
    """Format coords from ipyleaflet polygon  

    Args:
        polygon (dict): Polygon dict from ipyleaflet

    Returns:
        str: WKT style string
    """
# WKT geometry
    context = {'type' : 'Polygon', 'coordinates' : polygon['geometry']['coordinates']}
    geometry = shape(context).wkt
    reversed_poly = orient(geometry, sign=-1.0 if geometry.exterior.is_ccw else 1.0)
    geometry = geometry.replace("POLYGON ", "POLYGON").replace("MULTIPOLYGON ", "MULTIPOLYGON")
    geometry = geometry.replace("\n", "").replace("\r", "").strip()
    geometry = geometry.replace(") ,", "),").replace("( ", "(").replace(" )", ")")
    return geometry
```

```python
m = ipyleaflet.Map(center=[46.275,-72.2291], zoom=7)
#-72.2291, 46.275
shapes = []

def handle_draw(target, action, geo_json):
    global shapes
    shapes.append(geo_json)

draw_control = ipyleaflet.DrawControl()
draw_control.on_draw(handle_draw)
m.add_control(draw_control)
m
```

```python

config = read_config(config_path)
if shapes:
    for polygon in shapes:
        coords = get_coordinates(polygon)
    
    print(coords)
    if "gbif" in config.keys():
        print(True)
        config['gbif']['GEOMETRY'] = coords

    print(config)

write_config(config, config_path)
print(True)
        
```
