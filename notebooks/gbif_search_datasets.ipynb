```python
from pygbif import registry
from pprint import pprint
datasets = []
```

```python
query = "quebec"
response = registry.dataset_suggest(q=query, type="OCCURRENCE", publishingCountry= 'CA' )
for i, data in enumerate(response): 
    print("*"*100)
    print(i)
    pprint(data['title'])
```

Choose which dataset to add

```python
item = 0
out = response[item]
```

```python
pprint(out)
```

```python
datasets.append(out)
```

```python
for data in datasets:
    print(data['title'])

```

Output

```python
path = "/home/manat/projects/CScorer/datasets.yaml"
output = {'expert': {}}

for data in datasets:
    output['expert'][data['title']] = data
import yaml
try:
    file=open(path,"w")
    yaml.dump(output,file)
    file.close() 
except Exception as e:
    raise RuntimeError(e)
```
