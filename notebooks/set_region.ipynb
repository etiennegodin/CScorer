Quick notebook to help define region of interest (roi)

```python

import ipyleaflet
import yaml
from pathlib import Path
from shapely.geometry import shape
from pprint import pprint
import geopandas as gpd
from Cscorer.core import write_config

config = {}
config_out = {}

config["TAXON_KEY"] = 7707728
config['BASIS_OF_RECORD'] = "HUMAN_OBSERVATION"
config['HAS_COORDINATE'] = True
```

```python
def get_coordinates(polygon:dict):
# WKT geometry
    context = {'type' : 'Polygon', 'coordinates' : polygon['geometry']['coordinates']}
    geometry = shape(context).wkt
    geometry = geometry.replace("POLYGON ", "POLYGON").replace("MULTIPOLYGON ", "MULTIPOLYGON")
    geometry = geometry.replace("\n", "").replace("\r", "").strip()
    geometry = geometry.replace(") ,", "),").replace("( ", "(").replace(" )", ")")
    return geometry
```

```python
m = ipyleaflet.Map(center=[45.5, -73.6], zoom=13)
shapes = []

def handle_draw(target, action, geo_json):
    global shapes
    shapes.append(geo_json)

draw_control = ipyleaflet.DrawControl()
draw_control.on_draw(handle_draw)
m.add_control(draw_control)
m
```

```python
for polygon in shapes:
    coords = get_coordinates(polygon)
    config["GEOMETRY"] = coords

config_out['gbif'] = config

write_config(config_out, "/home/manat/projects/CScorer/work/dev/config.yaml")
    
```
