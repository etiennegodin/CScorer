```python
import requests
from urllib.parse import quote
from pprint import pprint
```

```python

# Convert to the special syntax
def fields_to_string(fields_dict, level=0):
    parts = []
    for key, value in fields_dict.items():
        if isinstance(value, dict):
            nested = fields_to_string(value, level + 1)
            parts.append(f"{key}:({nested})")
        elif value is True:
            parts.append(f"{key}:!t")
    return ','.join(parts)
```

```python
fields = {
        "count" : True,
        "controlled_attribute" : {"id":True, "label":True},
        "controlled_value" : {"id":True, "label":True},
        "month_of_year" : True

}
```

```python
fields_string = f"({fields_to_string(fields)})"
print(fields_string)
```

```python
params = {"taxon_id": "52833",
          "fields" : "(count:!t,controlled_attribute:(id:!t,label:!t),controlled_value:(id:!t,label:!t),month_of_year:!t)"
}
```

```python
try:
    response = requests.get('https://api.inaturalist.org/v2/observations/popular_field_values', params=params)
    print(response.url)
    response.raise_for_status()

    data = response.json()
    if data:
        res = data["results"]
    pprint(res)
except Exception as e:
    print(e)
```
