```python
import geopandas as gpd
import pandas as pd
from shapely import wkt
from pathlib import Path
```

```python
base_path = Path("/home/manat/projects/obsq/work/dev/data/inat/")
gpkgs = []
for i in range(1,10):
    gpkgs.append(base_path/f"iNaturalist_geomodel_Plantae_{i}.gpkg")
gpkgs
```

```python
poly = "POLYGON((-73.509156 45.524824, -74.519611 45.014526, -71.669469 45.014526, -70.203212 46.296853,-71.032444 46.821866, -71.444314 46.686379, -73.086303 46.045787, -73.509156 45.524824))"
bbox = wkt.loads(poly)
```

```python
gdf = gpd.GeoDataFrame()
for g in gpkgs:
    temp_gdf = gpd.read_file(g, bbox=bbox)
    pd.concat(gdf, temp_gdf)
```
