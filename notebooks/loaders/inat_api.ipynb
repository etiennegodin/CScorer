```python
import requests
from urllib.parse import quote
from pprint import pprint
```

```python

# Convert to the special syntax
def fields_to_string(fields_dict, level=0):
    parts = []
    for key, value in fields_dict.items():
        if isinstance(value, dict):
            nested = fields_to_string(value, level + 1)
            parts.append(f"{key}:({nested})")
        elif value is True:
            parts.append(f"{key}:!t")
    return ','.join(parts)
```

```python
fields = {'site_id' : True,
          "identifications_most_disagree" : True,
          "reviewed_by" : True}
```

```python
fields_string = f"({fields_to_string(fields)})"

params = {'uuid': "f6886166-241c-4b1f-8ed6-fbb39e6dd31e",
          'fields': fields_string}
print(fields_string)
```

```python
r = requests.get(f'https://api.inaturalist.org/v2/observations/', params=params)
print(r.url)
pprint(r.json())
```
