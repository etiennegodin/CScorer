```python
import requests

url = "https://api.inaturalist.org/v1/observations"

params = {
    "taxon_name": "Danaus plexippus",
    "per_page": 5,   # number of results per page (max 200)
    "page": 1
}

response = requests.get(url, params=params)
data = response.json()
print(data.keys())
pprint(data["results"][0].keys())  # explore structure
```

```python
def get_user_observations(username, per_page=100):
    url = "https://api.inaturalist.org/v1/observations"
    params = {"user_login": username, "per_page": per_page}
    res = requests.get(url, params=params).json()
    if res is not None:
        return res["results"]

obs_user = get_user_observations("egodin", per_page=10)
print([o["species_guess"] for o in obs_user])
```

```python
def get_user_data(user_login):
    url = f"https://api.inaturalist.org/v1/users/{user_login}"
    res = requests.get(url).json()
    user = res["results"][0]
    if user is not None:
        return user
user = get_user_data('egodin')
pprint(user)
```

Looping observations

```python
import requests
ids = [1000001, 1000002, 1000003]
url = f"https://api.inaturalist.org/v1/observations/{','.join(map(str, ids))}"

response = requests.get(url)
data = response.json()
observations = data["results"]

for obs in observations:
    print(f"Obs {obs['id']}: {obs['species_guess']} at {obs.get('place_guess')}")
```

Identification only

```python
url = "https://api.inaturalist.org/v1/identifications"
params = {"observation_id": ",".join(map(str, ids))}
res = requests.get(url, params=params).json()

for i in res["results"]:
    print(f"{i['user']['login']} identified it as {i['taxon']['name']}")
```
